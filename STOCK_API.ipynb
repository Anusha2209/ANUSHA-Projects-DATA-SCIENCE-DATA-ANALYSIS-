```python
#TO COLLECT REAL_TIME STOCK MARKET DATA USING "yfinance" API:
```

```python
import pandas as pd
import yfinance as yf
from datetime import datetime


```

```python
start_date=datetime.now() - pd.DateOffset(months=6)
end_date=datetime.now()
```

```python
tickers=['AAPL','MSFT','NFLX','GOOG']
```

```python
df_list=[]
```

```python
for ticker in tickers:
  data=yf.download(ticker,start=start_date,end=end_date)
  df_list.append(data)

```

```python
df=pd.concat(df_list,keys=tickers)
print(df.head())
```

```python
# IN THE ABOVE DATASET, DATECOLUMN IS THE INDEX SO BEFORE MOVING FURTHER WE NEED TO CHANGE IT
```

```python
df=df.reset_index()
print(df.head())
```

```python
#now lets look at the performance in the stock market
```

```python
import plotly.express as px
fig=px.line(df,x='Date',y='Close',color='level_0',title="Stock Market Performance for the last six months")
fig.show()
```

```python
fig=px.area(df,x='Date',y='Close',color='level_0',facet_col='level_0',labels={'Date':'Date','Close':'Closing Price','Ticker':'Company'},title="Stock Prices for Apple,Microsoft,Netflix, and Google")
fig.show()
```

```python
  #NOW LET US ANALYZE THE MOVING AVERAGES TO IDENTIFY TRENDS ANS PATTERNS IN EACH COMPANY"S STOCK PRICE MOVEMENTS OVER A PERIOD TIME
```

```python
df['MA10']=df.groupby('level_0')['Close'].rolling(window=10).mean().reset_index(0,drop=True)
df['MA20']=df.groupby('level_0')['Close'].rolling(window=20).mean().reset_index(0,drop=True)
```

```python
for level_0,group in df.groupby('level_0'):
  print(f'Moving Averages for {level_0}')
  print(group[['MA10','MA20']])
```

```python
# NOW LET US VISUALIZE MOVING AVERAGES OF THE FOLLOWING COMPANIES
```

```python
for level_0,group in df.groupby('level_0'):
  fig=px.line(group,x='Date',y=['Close','MA10','MA20'],title=f"{level_0}Moving Averages")
  fig.show()
```

```python
#WHEN MA10 CROSSES ABOVE MA20----> (BULLISH SIGNAL INDICATING THAT STOCKS PRICE  WILL CONTINUE TO RISE)
#WHEN MA10 CROSSES BELOW MA20----> (BEARISH SIGNAL STATES THAT STOCK PRICE WILL CONTINUE FALLING)
```

```python
#NOW LET US UNDERSTAAND THE VOLATILITY OF ALL COMPANIES
#  VOLATILITY---> MEASURE OF HOW MUCH AND HOW OFTEN THE STOCK PRICE OR MARKET FLUCTUATES OVER A GIVEN PERIOD OF TIME
#NOW LET US VISUALISE THE VOLATILITY OF ALL THE COMPANIES
```

```python
df['Volatility']=df.groupby('level_0')['Close'].pct_change().rolling(window=10).std().reset_index(0,drop=True)
fig=px.line(df,x='Date',y='Volatility',color='level_0',title='Volatility of All Companies')
fig.show()
```

```python
#HIGH VOLATILITY INDICATES THAT LARGE AND FREQUENT PRICE MOVEMENTS OF STOCK /MARKET EXPERIENCES
#LOW VOLATILITY INDICATES THAT MARKET EXPERIENCES SMALLER /LESS FREQUENT PRICE  MOVEMENTS
```

```python
#NOW LET US ANALYZE THE CORRELATION BETWEEN STOCK PRICES OF APPLE AND MICROSOFT
```

```python
#CREATE A DATAFRAME WITH STOCK PRICES OF APPLE AND MICROSOFT
apple=df.loc[df['level_0']=='AAPL',['Date','Close']].rename(columns={'Close': 'AAPL'})
microsoft=df.loc[df['level_0']=='MSFT',['Date','Close']].rename(columns={'Close':'MSFT'})
df_corr=pd.merge(apple,microsoft,on='Date')
```

```python
#CREATE A SCATTER PLOT TO VISUALIZE THE CORRELATION
fig=px.scatter(df_corr,x='AAPL',y='MSFT',trendline='ols',title='Correlation between Apple  and Microsoft')
fig.show()
```

```python
#FROM THE ABOVE PLOT WE CAN SEE THAT THERE IS A STRONG RALATIONSHIP BETWEEN STOCK PRICES OF APPLE AND MICROSOFT(REASON MAY BE DUE TO FACTORS  SUCH AS INDUSTRY TRENDS,MARKET CONDITIONS,OR COMMON BUSINESS PARTNERS OR CUSTOMERS)
```
